```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv('wb.csv')
```

```python
dataset
```

```python
dataset.drop(dataset.index[-5:], inplace=True)
```

```python
dataset[:].replace('..', np.nan, inplace=True)
dataset.dropna(subset=['Country Name' , 'Country Code', 'GDP per person employed (constant 2011 PPP $) [SL.GDP.PCAP.EM.KD]', 'Population, total [SP.POP.TOTL]' ,'Life expectancy at birth, total (years) [SP.DYN.LE00.IN]'], inplace=True)
```

```python
dataset
```

```python
dataset.columns=['T','TT','CN', 'CC', 'GDP', 'Pop', 'Life']
dataset
```

```python
dataset.Pop = pd.to_numeric(dataset.Pop).values
dataset.Life = pd.to_numeric(dataset.Life).values
dataset.GDP = pd.to_numeric(dataset.GDP).values
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x=dataset.GDP,y=dataset.Life, s=dataset.Pop/300000,edgecolors='White')
plt.show()
```
