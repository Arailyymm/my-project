```python
import folium
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('city.csv')
```

```python
df.columns=['Name','Pop','Region','X','Y']
df
```

```python
df.shape
```

```python
da=df.dropna(subset=['X', 'Y'],how='any')
da
```

```python
k_map=folium.Map(location=[43,66], tiles='Stamen Terrain',zoom_start=4.6)
```

```python
popul=folium.map.FeatureGroup()


latitudes=list(da.Y)
longitudes=list(da.X)
labels=list(da.Pop)


for lat,lng, label in zip(latitudes,longitudes,labels):
    folium.Marker([lat,lng], popup=label).add_to(k_map)


k_map
```
