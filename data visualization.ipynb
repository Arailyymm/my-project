```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df =  pd.read_csv('olympics.csv')
cols = ['Country']
df.columns = cols
```

```python
df = pd.DataFrame(df.Country.str.split(';',1).tolist(),
                                   columns = ['Country','Medals'])
```
